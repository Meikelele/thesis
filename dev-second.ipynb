```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
import optuna

# Drzewo decyzyjne bez optymalizacji
tree_no_opt = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_no_opt.fit(X_train, y_train)
y_pred_no_opt = tree_no_opt.predict(X_test)
tree_probs_no_opt = tree_no_opt.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą Optuny
def objective(trial):
    max_depth = trial.suggest_int('max_depth', 3, 10000)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10000)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10000)

    model = DecisionTreeClassifier(
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10000)
best_params_optuna = study.best_params

tree_optuna = DecisionTreeClassifier(**best_params_optuna, random_state=42)
tree_optuna.fit(X_train, y_train)
y_pred_optuna = tree_optuna.predict(X_test)
tree_probs_optuna = tree_optuna.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą RandomizedSearchCV
param_dist = {
    'max_depth': range(3, 10000),
    'min_samples_split': range(2, 10000),
    'min_samples_leaf': range(1, 10000)
}
random_search = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42), param_distributions=param_dist,
    n_iter=10000, scoring='roc_auc', random_state=42, cv=3, n_jobs=-1
)
random_search.fit(X_train, y_train)
best_params_random = random_search.best_params_

tree_random = DecisionTreeClassifier(**best_params_random, random_state=42)
tree_random.fit(X_train, y_train)
y_pred_random = tree_random.predict(X_test)
tree_probs_random = tree_random.predict_proba(X_test)[:, 1]

# Obliczenie metryk dla każdego modelu
def evaluate_model(y_true, y_pred, y_proba, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_proba)
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc_score:.4f}")

    return auc_score

auc_no_opt = evaluate_model(y_test, y_pred_no_opt, tree_probs_no_opt, "Decision Tree (No Optimization)")
confusion(y_test, y_pred_no_opt, "Decision Tree (No Optimization)")

auc_optuna = evaluate_model(y_test, y_pred_optuna, tree_probs_optuna, "Decision Tree (Optuna)")
confusion(y_test, y_pred_optuna, "Decision Tree (Optuna)")

auc_random = evaluate_model(y_test, y_pred_random, tree_probs_random, "Decision Tree (RandomSearch)")
confusion(y_test, y_pred_random, "Decision Tree (RandomSearch)")

# Wykres ROC
fpr_no_opt, tpr_no_opt, _ = roc_curve(y_test, tree_probs_no_opt)
fpr_optuna, tpr_optuna, _ = roc_curve(y_test, tree_probs_optuna)
fpr_random, tpr_random, _ = roc_curve(y_test, tree_probs_random)

plt.figure(figsize=(8, 6))
plt.plot(fpr_no_opt, tpr_no_opt, marker='.', label=f'No Optimization (AUROC = {auc_no_opt:.3f})')
plt.plot(fpr_optuna, tpr_optuna, marker='.', label=f'Optuna (AUROC = {auc_optuna:.3f})')
plt.plot(fpr_random, tpr_random, marker='.', label=f'RandomSearch (AUROC = {auc_random:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Rozdział 2 - RandomForest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
import optuna

# Random Forest bez optymalizacji
rf_no_opt = RandomForestClassifier(max_features=5, n_estimators=500, random_state=42)
rf_no_opt.fit(X_train, y_train)
y_pred_no_opt = rf_no_opt.predict(X_test)
rf_probs_no_opt = rf_no_opt.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą Optuny
def objective(trial):
    max_depth = trial.suggest_int('max_depth', 3, 1000)
    n_estimators = trial.suggest_int('n_estimators', 100, 10000)
    max_features = trial.suggest_int('max_features', 1, X_train.shape[1])
    min_samples_split = trial.suggest_int('min_samples_split', 2, 1000)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 1000)

    model = RandomForestClassifier(
        max_depth=max_depth,
        n_estimators=n_estimators,
        max_features=max_features,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1500)
best_params_optuna = study.best_params

rf_optuna = RandomForestClassifier(**best_params_optuna, random_state=42)
rf_optuna.fit(X_train, y_train)
y_pred_optuna = rf_optuna.predict(X_test)
rf_probs_optuna = rf_optuna.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą RandomizedSearchCV
param_dist = {
    'max_depth': range(3, 1000),
    'n_estimators': range(100, 1000),
    'max_features': range(1, X_train.shape[1] + 1),
    'min_samples_split': range(2, 1000),
    'min_samples_leaf': range(1, 1000)
}
random_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42), param_distributions=param_dist,
    n_iter=1500, scoring='roc_auc', random_state=42, cv=3, n_jobs=-1
)
random_search.fit(X_train, y_train)
best_params_random = random_search.best_params_

rf_random = RandomForestClassifier(**best_params_random, random_state=42)
rf_random.fit(X_train, y_train)
y_pred_random = rf_random.predict(X_test)
rf_probs_random = rf_random.predict_proba(X_test)[:, 1]

# Obliczenie metryk dla każdego modelu
def evaluate_model(y_true, y_pred, y_proba, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_proba)
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc_score:.4f}")

    return auc_score

auc_no_opt = evaluate_model(y_test, y_pred_no_opt, rf_probs_no_opt, "Random Forest (No Optimization)")
confusion(y_test, y_pred_no_opt, "Random Forest (No Optimization)")

auc_optuna = evaluate_model(y_test, y_pred_optuna, rf_probs_optuna, "Random Forest (Optuna)")
confusion(y_test, y_pred_optuna, "Random Forest (Optuna)")

auc_random = evaluate_model(y_test, y_pred_random, rf_probs_random, "Random Forest (RandomSearch)")
confusion(y_test, y_pred_random, "Random Forest (RandomSearch)")

# Wykres ROC
fpr_no_opt, tpr_no_opt, _ = roc_curve(y_test, rf_probs_no_opt)
fpr_optuna, tpr_optuna, _ = roc_curve(y_test, rf_probs_optuna)
fpr_random, tpr_random, _ = roc_curve(y_test, rf_probs_random)

plt.figure(figsize=(8, 6))
plt.plot(fpr_no_opt, tpr_no_opt, marker='.', label=f'No Optimization (AUROC = {auc_no_opt:.3f})')
plt.plot(fpr_optuna, tpr_optuna, marker='.', label=f'Optuna (AUROC = {auc_optuna:.3f})')
plt.plot(fpr_random, tpr_random, marker='.', label=f'RandomSearch (AUROC = {auc_random:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Rozdział 3 - XGBoost

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
import optuna
from xgboost import XGBClassifier

# XGBoost bez optymalizacji
xgb_no_opt = XGBClassifier(random_state=42)
xgb_no_opt.fit(X_train, y_train)
y_pred_no_opt = xgb_no_opt.predict(X_test)
xgb_probs_no_opt = xgb_no_opt.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą Optuny
def objective(trial):
    max_depth = trial.suggest_int('max_depth', 3, 1000)
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    learning_rate = trial.suggest_float('learning_rate', 0.01, 3.0)
    colsample_bytree = trial.suggest_float('colsample_bytree', 0.3, 100.0)
    subsample = trial.suggest_float('subsample', 0.3, 100.0)

    model = XGBClassifier(
        max_depth=max_depth,
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        colsample_bytree=colsample_bytree,
        subsample=subsample,
        random_state=42
    )
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1500)
best_params_optuna = study.best_params

xgb_optuna = XGBClassifier(**best_params_optuna, random_state=42)
xgb_optuna.fit(X_train, y_train)
y_pred_optuna = xgb_optuna.predict(X_test)
xgb_probs_optuna = xgb_optuna.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą RandomizedSearchCV
param_dist = {
    'max_depth': range(3, 1000),
    'n_estimators': range(100, 1000),
    'learning_rate': np.linspace(0.01, 0.3, 10),
    'colsample_bytree': np.linspace(0.3, 1.0, 10),
    'subsample': np.linspace(0.3, 1.0, 10)
}
random_search = RandomizedSearchCV(
    XGBClassifier(random_state=42), param_distributions=param_dist,
    n_iter=1500, scoring='roc_auc', random_state=42, cv=3, n_jobs=-1
)
random_search.fit(X_train, y_train)
best_params_random = random_search.best_params_

xgb_random = XGBClassifier(**best_params_random, random_state=42)
xgb_random.fit(X_train, y_train)
y_pred_random = xgb_random.predict(X_test)
xgb_probs_random = xgb_random.predict_proba(X_test)[:, 1]

# Obliczenie metryk dla każdego modelu
def evaluate_model(y_true, y_pred, y_proba, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_proba)
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc_score:.4f}")

    return auc_score

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Non Hepatitis', 'Hepatitis'], yticklabels=['Non Hepatitis', 'Hepatitis'])
    plt.title(f"Confusion Matrix: {title}")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.show()

auc_no_opt = evaluate_model(y_test, y_pred_no_opt, xgb_probs_no_opt, "XGBoost (No Optimization)")
plot_confusion_matrix(y_test, y_pred_no_opt, "XGBoost (No Optimization)")

auc_optuna = evaluate_model(y_test, y_pred_optuna, xgb_probs_optuna, "XGBoost (Optuna)")
plot_confusion_matrix(y_test, y_pred_optuna, "XGBoost (Optuna)")

auc_random = evaluate_model(y_test, y_pred_random, xgb_probs_random, "XGBoost (RandomSearch)")
plot_confusion_matrix(y_test, y_pred_random, "XGBoost (RandomSearch)")

# Wykres ROC
fpr_no_opt, tpr_no_opt, _ = roc_curve(y_test, xgb_probs_no_opt)
fpr_optuna, tpr_optuna, _ = roc_curve(y_test, xgb_probs_optuna)
fpr_random, tpr_random, _ = roc_curve(y_test, xgb_probs_random)

plt.figure(figsize=(8, 6))
plt.plot(fpr_no_opt, tpr_no_opt, marker='.', label=f'No Optimization (AUROC = {auc_no_opt:.3f})')
plt.plot(fpr_optuna, tpr_optuna, marker='.', label=f'Optuna (AUROC = {auc_optuna:.3f})')
plt.plot(fpr_random, tpr_random, marker='.', label=f'RandomSearch (AUROC = {auc_random:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Rozdział 9 - MLP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
import optuna

# MLP bez optymalizacji
mlp_no_opt = MLPClassifier(random_state=42, max_iter=200)
mlp_no_opt.fit(X_train, y_train)
y_pred_no_opt = mlp_no_opt.predict(X_test)
mlp_probs_no_opt = mlp_no_opt.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą Optuny
def objective(trial):
    hidden_layer_sizes = (trial.suggest_int('hidden_layer_neurons', 10, 1000),)
    alpha = trial.suggest_float('alpha', 1e-10, 1e-1, log=True)
    learning_rate_init = trial.suggest_float('learning_rate_init', 1e-10, 1e-1, log=True)
    activation = trial.suggest_categorical('activation', ['identity', 'logistic', 'tanh', 'relu'])
    solver = trial.suggest_categorical('solver', ['lbfgs', 'adam']) 

    model = MLPClassifier(
        hidden_layer_sizes=hidden_layer_sizes,
        alpha=alpha,
        learning_rate_init=learning_rate_init,
        activation=activation,
        solver=solver,
        random_state=42,
        max_iter=1500
    )
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1500)
best_params_optuna = study.best_params

mlp_optuna = MLPClassifier(
    hidden_layer_sizes=(best_params_optuna['hidden_layer_neurons'],),
    alpha=best_params_optuna['alpha'],
    learning_rate_init=best_params_optuna['learning_rate_init'],
    activation=best_params_optuna['activation'],
    solver=best_params_optuna['solver'],
    random_state=42,
    max_iter=1500
)
mlp_optuna.fit(X_train, y_train)
y_pred_optuna = mlp_optuna.predict(X_test)
mlp_probs_optuna = mlp_optuna.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą RandomizedSearchCV
param_dist = {
    'hidden_layer_sizes': [(x,) for x in range(10, 101, 10)],
    'alpha': np.logspace(-10, -1, 10),
    'learning_rate_init': np.logspace(-10, -1, 10),
    'activation': ['identity', 'logistic', 'tanh', 'relu'],
    'solver': ['lbfgs', 'adam']
}
random_search = RandomizedSearchCV(
    MLPClassifier(random_state=42, max_iter=1500), 
    param_distributions=param_dist,
    n_iter=1500, scoring='roc_auc', random_state=42, cv=3, n_jobs=-1
)
random_search.fit(X_train, y_train)
best_params_random = random_search.best_params_

mlp_random = MLPClassifier(**best_params_random, random_state=42, max_iter=1500)
mlp_random.fit(X_train, y_train)
y_pred_random = mlp_random.predict(X_test)
mlp_probs_random = mlp_random.predict_proba(X_test)[:, 1]

# Funkcja do oceny modelu
def evaluate_model(y_true, y_pred, y_proba, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_proba)
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc_score:.4f}")
    return auc_score

# Ocena modeli
auc_no_opt = evaluate_model(y_test, y_pred_no_opt, mlp_probs_no_opt, "MLP (No Optimization)")
confusion(y_test, y_pred_no_opt, "MLP (No Optimization)")

auc_optuna = evaluate_model(y_test, y_pred_optuna, mlp_probs_optuna, "MLP (Optuna)")
confusion(y_test, y_pred_optuna, "MLP (Optuna)")

auc_random = evaluate_model(y_test, y_pred_random, mlp_probs_random, "MLP (RandomSearch)")
confusion(y_test, y_pred_random, "MLP (RandomSearch)")

# Wykres ROC
fpr_no_opt, tpr_no_opt, _ = roc_curve(y_test, mlp_probs_no_opt)
fpr_optuna, tpr_optuna, _ = roc_curve(y_test, mlp_probs_optuna)
fpr_random, tpr_random, _ = roc_curve(y_test, mlp_probs_random)

plt.figure(figsize=(8, 6))
plt.plot(fpr_no_opt, tpr_no_opt, marker='.', label=f'No Optimization (AUROC = {auc_no_opt:.3f})')
plt.plot(fpr_optuna, tpr_optuna, marker='.', label=f'Optuna (AUROC = {auc_optuna:.3f})')
plt.plot(fpr_random, tpr_random, marker='.', label=f'RandomSearch (AUROC = {auc_random:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Rozdział 10 - NuSVC

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import NuSVC
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
import optuna

# NuSVC bez optymalizacji
nusvc_no_opt = NuSVC(kernel='rbf', probability=True, random_state=42, nu=0.5, gamma='scale')
nusvc_no_opt.fit(X_train, y_train)
y_pred_no_opt = nusvc_no_opt.predict(X_test)
nusvc_probs_no_opt = nusvc_no_opt.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą Optuny
def objective(trial):
    nu = trial.suggest_float('nu', 0.001, 0.999, log=False)
    gamma = trial.suggest_float('gamma', 1e-10, 1e-1, log=True)
    
    model = NuSVC(kernel='rbf', probability=True, nu=nu, gamma=gamma, random_state=42)
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1500)
best_params_optuna = study.best_params

nusvc_optuna = NuSVC(kernel='rbf', probability=True, random_state=42, **best_params_optuna)
nusvc_optuna.fit(X_train, y_train)
y_pred_optuna = nusvc_optuna.predict(X_test)
nusvc_probs_optuna = nusvc_optuna.predict_proba(X_test)[:, 1]

# Optymalizacja za pomocą RandomizedSearchCV
param_dist = {
    'nu': np.linspace(0.01, 0.99, 500),           
    'gamma': np.logspace(-10, -1, 500)           
}
random_search = RandomizedSearchCV(
    NuSVC(kernel='rbf', probability=True, random_state=42), 
    param_distributions=param_dist,
    n_iter=1500, scoring='roc_auc', random_state=42, cv=3, n_jobs=-1
)
random_search.fit(X_train, y_train)
best_params_random = random_search.best_params_

nusvc_random = NuSVC(kernel='rbf', probability=True, random_state=42, **best_params_random)
nusvc_random.fit(X_train, y_train)
y_pred_random = nusvc_random.predict(X_test)
nusvc_probs_random = nusvc_random.predict_proba(X_test)[:, 1]

# Funkcja do oceny modelu
def evaluate_model(y_true, y_pred, y_proba, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_proba)
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"AUC: {auc_score:.4f}")
    return auc_score

# Ocena modeli
auc_no_opt = evaluate_model(y_test, y_pred_no_opt, nusvc_probs_no_opt, "NuSVC (No Optimization)")
confusion(y_test, y_pred_no_opt, "NuSVC (No Optimization)")

auc_optuna = evaluate_model(y_test, y_pred_optuna, nusvc_probs_optuna, "NuSVC (Optuna)")
confusion(y_test, y_pred_optuna, "NuSVC (Optuna)")

auc_random = evaluate_model(y_test, y_pred_random, nusvc_probs_random, "NuSVC (RandomSearch)")
confusion(y_test, y_pred_random, "NuSVC (RandomSearch)")

# Wykres ROC
fpr_no_opt, tpr_no_opt, _ = roc_curve(y_test, nusvc_probs_no_opt)
fpr_optuna, tpr_optuna, _ = roc_curve(y_test, nusvc_probs_optuna)
fpr_random, tpr_random, _ = roc_curve(y_test, nusvc_probs_random)

plt.figure(figsize=(8, 6))
plt.plot(fpr_no_opt, tpr_no_opt, marker='.', label=f'No Optimization (AUROC = {auc_no_opt:.3f})')
plt.plot(fpr_optuna, tpr_optuna, marker='.', label=f'Optuna (AUROC = {auc_optuna:.3f})')
plt.plot(fpr_random, tpr_random, marker='.', label=f'RandomSearch (AUROC = {auc_random:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
